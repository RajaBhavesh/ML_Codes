```python
df
```

```python
plt.xlabel('Area(sqft)')
plt.ylabel('Price(US$)')
plt.scatter(df.area,df.price,color='red',marker='+')
plt.plot()
```

```python
#now we can use linear regreesion model as e can see that it ids good to use taht model here
reg=linear_model.LinearRegression()#createad a object
reg.fit(df[['area']],df.price)#now i am fitting the model
```

```python
#now our model is trained and ready to predict
reg.predict([[3300]])
```

```python
#to check m and c values
reg.coef_
```
