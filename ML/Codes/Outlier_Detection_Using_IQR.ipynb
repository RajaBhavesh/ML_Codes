```python
import pandas as pd
```

```python
df=pd.read_csv('heights (1).csv')
```

```python
df.head()
```

```python
df.height.describe()
```

```python
Q1=df.height.quantile(0.25)
Q2=df.height.quantile(0.50)
Q3=df.height.quantile(0.75)
```

```python
IQR=Q3-Q1
IQR
```

```python
upper_limit=Q3+1.5*IQR
lower_limit=Q1-1.5*IQR
```

```python
lower_limit
```

```python
upper_limit
```

```python
df[(df['height']<lower_limit)|(df['height']>upper_limit)] #these are outliers
```

```python
newdf=df[(df['height']>=lower_limit)&(df['height']<=upper_limit)] #these are outliers
```

```python
newdf
```
