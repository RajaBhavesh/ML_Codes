```python
import pandas as pd
df=pd.read_csv('spam.csv')
```

```python
df.head()
```

```python
df.groupby('Category').describe()
```

```python
#ml model understand numbers
#1-> it is very esy to convert sapam and ham to uumbers
df['spam']=df.Category.apply(lambda x: 1 if x=='spam' else 0)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_tarin,y_test=train_test_split(df.Message,df.spam,test_size=0.25)
```

```python
#stil we have message columns a stext now how to convert this into numerical ??
```

```python
#we'll use CountVectorizer Method
from sklearn.feature_extraction.text import CountVectorizer
v=CountVectorizer()
X_train_count=v.fit_transform(X_train.values)
```

```python
X_train_count.toarray()[:3]
```

```python
#now we are going to import model
from sklearn.naive_bayes import MultinomialNB
```

```python
model=MultinomialNB()
model.fit(X_train_count,y_tarin)
```

```python
#checking model
email=['Hey, mohan can we go for playing cricket ?','Upto 20% discount on parking, exclusive offer just for you']
```

```python
email_count=v.transform(email)
```

```python
model.predict(email_count)
```

```python
X_test_count=v.transform(X_test)
```

```python
model.score(X_test_count,y_test)
```

```python
from sklearn.pipeline import Pipeline
```

```python
clf=Pipeline([('vectorizer',CountVectorizer()),('nb',MultinomialNB())])
```

```python
clf.fit(X_train,y_tarin)
```

```python
clf.predict(email)
```

```python
clf.score(X_test,y_test)
```
