```python
import pandas as pd
df=pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df=df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
```

```python
df.head()
```

```python
target=df['Survived']
inputs=df.drop(['Survived'],axis='columns')
```

```python
dummies=pd.get_dummies(df.Sex)
dummies
```

```python
inputs=pd.concat([inputs,dummies],axis='columns')
```

```python
inputs.head()
```

```python
inputs=inputs.drop(['Sex'],axis='columns')
```

```python
inputs.head()
```

```python
inputs.info()
```

```python
inputs[inputs.Age.isnull()]
```

```python
mm=inputs.Age.mean()
```

```python
mm
```

```python
inputs.Age=inputs.Age.fillna(mm)
```

```python
inputs.info()
```

```python
#now we are going to split into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(inputs,target,test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
len(inputs)
```

```python
#now there are many naive base classes but here we are going to import GaussianNB
from sklearn.naive_bayes import GaussianNB
```

```python
model=GaussianNB()
```

```python
#to train the model
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
