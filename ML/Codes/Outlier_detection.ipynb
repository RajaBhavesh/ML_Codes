```python
import pandas as pd
df=pd.read_csv('heights.csv')
```

```python
df.head()
```

```python
#now for outlier i am considering 0.95
max_threshold=df['height'].quantile(0.95)
max_threshold
```

```python
df[df['height']>max_threshold] #outlier
```

```python
min_threshold=df['height'].quantile(0.05)
min_threshold
```

```python
df[df['height']<min_threshold] #outlier
```

```python
df=df[(df['height']>min_threshold)&(df['height']<max_threshold)]
```
