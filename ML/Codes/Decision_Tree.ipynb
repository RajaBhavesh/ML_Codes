```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('salaries.csv')
```

```python
df.head()
```

```python
inputs=df.drop('salary_more_then_100k',axis='columns')
```

```python
inputs.head()
```

```python
target=df['salary_more_then_100k']
```

```python
target.head()
```

```python
#as we now need to convert our text data to numerical data
#1 -> use pandas.get_dummies which we already discussed
#2-> using labelEncoder
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
inputs['company_n']=le_company.fit_transform(inputs['company'])
inputs['job_n']=le_job.fit_transform(inputs['job'])
inputs['degree_n']=le_degree.fit_transform(inputs['degree'])
```

```python
inputs.head()
```

```python
input_n=inputs.drop(['company','job','degree'],axis='columns')
```

```python
input_n.head()
```

```python
#now we are ready to train or classifieer
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(input_n,target)
```

```python
model.score(input_n,target)
```

```python
model.predict([[2,0,1]])
```

```python
import joblib
```

```python
joblib.dump(model,'my_model')
```

```python
from google.colab import files
files.download('my_model')
```

```python
#now anyone can use
raja=joblib.load('my_model')
```

```python
raja.predict([[2,0,1]])
```

```python
df=pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df=df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
```

```python
df.head()
```

```python
inputs=df.drop(['Survived'],axis='columns')
```

```python
target=df['Survived']
```

```python
inputs['Sex']=inputs['Sex'].map({'male':1,'female':2})
```

```python
inputs.info()
```

```python
inputs['Age']=inputs['Age'].fillna(inputs['Age'].mean())
```

```python
inputs.info()
```

```python
#now we are done with eda, now go for model
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(inputs,target,test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
