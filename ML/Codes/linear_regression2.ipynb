```python
df=pd.read_csv('hiring.csv')
```

```python
df
```

```python
df['experience']=df['experience'].fillna(0)
df['experience']=df['experience'].replace({'one':1,'two':2,'three':3,'four':4,'five':5,'six':6,'seven':7,'eight':8,'ten':10,'eleven':11})
```

```python
df['experience'].astype(int)
```

```python
df['experience'].dtype
```

```python
df.info()
```

```python
df['test_score(out of 10)']=df['test_score(out of 10)'].fillna(df['test_score(out of 10)'].median())
```

```python
df
```

```python
#now we have done with data processing
#let's do ml modeling
reg=linear_model.LinearRegression()
```

```python
#created a object
reg.fit(df[['experience','test_score(out of 10)','interview_score(out of 10)']],df['salary($)'])
```

```python
#now we can predict
reg.predict([[2,9,6]])
```

```python
reg.predict([[12,10,10]])
```
